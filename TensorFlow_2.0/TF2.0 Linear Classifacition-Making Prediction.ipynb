```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(
    data.data,data.target,test_size=0.33)
```

```python
N,D=X_train.shape
```

```python
N,D
```

```python
# StandardScaler kullanılarak veri ölçeklendirme

# 1. StandardScaler nesnesi oluşturuluyor
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()  # Veriler 0 ortalama ve 1 standart sapma ile ölçeklenir.

# 2. Eğitim verilerini ölçeklendiriyoruz
X_train = scaler.fit_transform(X_train)  # Eğitim verisinin ortalama ve standart sapması hesaplanır ve ölçeklenir.

# 3. Test verilerini ölçeklendiriyoruz
X_test = scaler.transform(X_test)  # Test verisi, yalnızca eğitimde hesaplanan parametrelerle dönüştürülür.
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(D,)),  # Giriş katmanı (özellik sayısı D)
    tf.keras.layers.Dense(1, activation='sigmoid')  # Çıkış katmanı (sigmoid aktivasyon)
])
```

```python
# Alternative
# model = tf.keras.models.Sequential()
# model.add(tf.keras.layers.Dense(1,input_shape=(D,),activation='sigmoid'))
```

```python
#compilde the model

# Modelin Derlenmesi
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Modelin Eğitimi
r = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100)
```

```python
# Model Performansını Değerlendirme
print("Train score:", model.evaluate(X_train, y_train))
print("Test score:", model.evaluate(X_test, y_test))
```

```python
# Kayıp Grafiği
import matplotlib.pyplot as plt
plt.plot(r.history['loss'], label='loss')  # Eğitim kayıpları
plt.plot(r.history['val_loss'], label='val_loss')  # Doğrulama kayıpları
plt.legend()
```

```python
# Doğruluk Grafiği
plt.plot(r.history['accuracy'], label='acc')  # Eğitim doğrulukları
plt.plot(r.history['val_accuracy'], label='val_acc')  # Doğrulama doğrulukları
plt.legend()
```

# Part 2: Making Predictions


```python
# make predictions
P = model.predict(X_test)
print(P) # they are outputs of the sigmoid,interpreted as probabilities p(y=1|x)
```

```python
#Round to get the actual predictions
#Note: has to be flattened since the targets are size (N,) while the predictions are size (N,1)
import numpy as np
P = np.round(P).flatten()
print(P)
```

```python
#Calculate the accuracy, copare it to evaluate() output
print("Manually calculated accuracy:", np.mean(P ==y_test))
print("Evaluate output:",model.evaluate(X_test,y_test))
```
