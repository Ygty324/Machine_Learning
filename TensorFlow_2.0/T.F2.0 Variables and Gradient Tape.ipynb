```python
# L(w) = w**2

w = tf.Variable(5.0)

def get_loss(w):
  return w**2

def get_grad(w):
  with tf.GradientTape() as tape:
    L = get_loss(w)
  g= tape.gradient(L, w)
  return g

optimizer = tf.keras.optimizers.SGD(learning_rate=0.1)

losses = []

for i in range(50):
  g = get_grad(w)
  optimizer.apply_gradients(zip([g], [w]))
  losses.append(get_loss(w))
```

```python
import matplotlib.pyplot as plt
plt.plot(losses)
print(f"Final loss: {get_loss(w)}" )
```

```python
w = tf.Variable(5.)
losses2 = []
for i in range(50):
  w = w-0.1*2*w
  losses2.append(w ** 2)
```

```python
plt.plot(losses,label='losses tf')
plt.plot(losses2,label='losses numpy')
plt.legend()
```
